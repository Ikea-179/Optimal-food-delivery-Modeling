```python
import json
import copy
import math
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
json_file = open('/Users/alan/Desktop/graph.json')
json_str = json_file.read()
points_graph = json.loads(json_str)

r_points = {
    '3': 'R1',
    '7': 'R2',
    '4': 'R3'
}

c_points = {
    '10': 'C1',
    '11': 'C2',
    '12': 'C3'
}

most_orders = 2
# Use recursion to solve the dynamic programming problem
def point_travel(point_current, distance_current, orders_get, orders_finish, travel_history):
    """
    point_current: the current position, 
    distance_current: the distance the rider has taken, 
    orders_get: delivery food in the cargo,     
    orders_finish: orders that have been finished, 
    travel_history: the record of the paths
    """
    # check if all the orders are finished
    if len(orders_finish) == len(c_points):
        if distance_current in distance_travel:
            distance_travel[distance_current].append(travel_history)
        else:
            distance_travel[distance_current] = [travel_history]
        return None
    # When calculating a distance (starting a new recursive call), only the travelling route distance smaller than the current distance is considered
    '''
    This section is very important for the prune of the recursion tree of the dynamic programming because it grows exponentially.
    '''
    if len(distance_travel) > 0:
        if distance_current >= min(distance_travel.keys()) - 1:
            return None
    # next step
    for key, value in points_graph[point_current].items():
        # When the state of the going to the next step point already exists in the travel, continue (avoiding backtracking)
        if (key, orders_get, orders_finish) in travel_history:
            continue
        distances_ = distance_current
        distances_ += value
        travel_history_ = travel_history.copy()
        orders_get_ = orders_get.copy()
        if key in r_points:
            # current point is a restaurant which has an order
            if len(orders_get) < most_orders and r_points[key] not in orders_get and r_points[key].replace('R', 'C') not in orders_finish:
                '''
                At the restaurant, if the cargo is not full, and if the current restaurant's order is not in the cargo, 
                choose whether to take the order or not (two options, core of dynamic programming)
                '''
                for i in range(2):
                    travel_history_ = travel_history.copy()
                    if i:
                        orders_get_.add(r_points[key])
                    travel_history_.append((key, orders_get_, orders_finish))
                    point_travel(key, distances_, orders_get_, orders_finish, travel_history_)
            else:
                # if we do not take this order into the cargo
                travel_history_.append((key, orders_get_, orders_finish))
                point_travel(key, distances_, orders_get_, orders_finish, travel_history_)
        elif key in c_points:
            # if the current point is customer
            order = c_points[key].replace('C', 'R')
            orders_finish_ = orders_finish.copy()
            # we finish the order if we have delivered it to the customer
            if order in orders_get:
                orders_get_.remove(order)
                orders_finish_.add(c_points[key])
            travel_history_.append((key, orders_get_, orders_finish_))
            point_travel(key, distances_, orders_get_, orders_finish_, travel_history_)
        else:
            # update the travel history
            travel_history_.append((key, orders_get, orders_finish))
            # start the recursive call of the adjacent vertex
            point_travel(key, distances_, orders_get, orders_finish, travel_history_)
    return None


if __name__ == '__main__':
    distance_travel = {}
    point_travel('2', 0, set(), set(), [])
    min_distance = min(distance_travel.keys())
    print(min_distance)
    print(len(distance_travel))
    for travel in distance_travel[min_distance]:
        print(travel)
```

```python
'''
This section is to generate the All-Pair-Shortest-Path Matrix (size 120 * 120) using the Floyd-Warshall Algorithm
'''


json_file = open('/Users/alan/Desktop/graph2.json')
json_str = json_file.read()
points_graph = json.loads(json_str)

nodes = []
dis = []
for key in points_graph.keys():
    nodes.append(key)
    dis.append([])

for i in range(len(nodes)):
    for j in range(len(nodes)):
        dis[i].append(math.inf)

for i in range(len(nodes)):
    for j in range(len(nodes)):
        if i == j:
            dis[i][j] = 0

for key1 in points_graph.keys():
    for key2 in points_graph[key1].keys():
        dis[nodes.index(key1)][nodes.index(key2)] = points_graph[key1][key2]

node_num = len(nodes)
count = 0
for i in range(node_num):
    for j in range(node_num):
        for k in range(node_num):
            dis[j][k] = min(dis[j][k], dis[j][i] + dis[i][k])

dis = np.array(dis)
```

```python
samples = random.sample(range(1, 121), 80)
```

```python
restaurants = samples[0: 40]
communities = samples[40:]
```

```python
random.seed(0)
```

```python
# pseudo-OOP
rider = {}
for i in range(1, 101):
    rider['P' + str(i)] = {'rider_number': i, 'cur_point': communities[random.randint(0, 39)], 'cargo': [], 
                           'start_cur_delivery_time': -1,
                           'finish_cur_delivery_time': -1, 
                           'credit_score': 10}
```

```python
# generate all the orders and the destinations for every minute
order_r = {}
order_c = {}
j = 0
for i in range(1, 181):
    order_r[i] = {}
    order_c[i] = {}
    older_j = j
    j += int(random.randint(20, 40))
    for k in range(older_j + 1, j + 1):
        order_r[i]['R_' + str(i) + "_" + str(k)] = restaurants[random.randint(0, 39)]
        order_c[i]['C_' + str(i) + "_" + str(k)] = communities[random.randint(0, 39)]
```

```python
awaiting_order_r = {}
awaiting_order_c = {}
finished_orders = {}
canceled_order_r = {}
canceled_order_c = {}
total_revenue = 0
speed = 0.5

for i in range(1, 181):
    # ------- check finished orders and delete them in the rider's cargo -------   
    for j in range(0, 100):
        keys = list(rider.keys())
        if i >= rider[keys[j]]['finish_cur_delivery_time']:
            rider[keys[j]]['cargo'] = []
        else:
            continue
            
#     ------- check awaiting orders -------   
    awaiting_order_key = list(awaiting_order_r.keys())
    for key in awaiting_order_key:
        if i - key >= 10:
            canceled_order_r[i] = awaiting_order_r.pop(key)
            canceled_order_c[i] = awaiting_order_c.pop(key)
            if (len(canceled_order_r[i]) > 15):
                # if there are too many canceled orders, the nubmer of orders in the incoming minute will decrease by 5.
                for number in range(0, 10):
                    if i < 180:
                        order_r[i + 1].popitem()
                        order_c[i + 1].popitem()
        else:
            for j in range(0, 100):
                rider = dict(sorted(rider.items(), key = lambda d: d[1]['credit_score'], reverse = True))
                keys = list(rider.keys())
                if i >= rider[keys[j]]['finish_cur_delivery_time']:
                    if len(rider[keys[j]]['cargo']) < 2:
                        awaiting_order_r_keys = list(awaiting_order_r[key].keys())
                        awaiting_order_c_keys = list(awaiting_order_c[key].keys())
                        if len(awaiting_order_r_keys) > 1:
                            for k in range(0, 2 - len(rider[keys[j]]['cargo'])):
                                rider[keys[j]]['cargo'].append({awaiting_order_r_keys[k]: awaiting_order_r[key].pop(awaiting_order_r_keys[k]), 
                                                                awaiting_order_c_keys[k]: awaiting_order_c[key].pop(awaiting_order_c_keys[k])})
                                rider[keys[j]]['start_cur_delivery_time'] = i
                        elif len(awaiting_order_r_keys) == 1:
                            rider[keys[j]]['cargo'].append({awaiting_order_r_keys[0]: awaiting_order_r[key].pop(awaiting_order_r_keys[0]), 
                                                            awaiting_order_c_keys[0]: awaiting_order_c[key].pop(awaiting_order_c_keys[0])})
                        else:
                            continue
                else:
                    continue
            
    # ------- assigning orders -------
    for j in range(0, 100):
        rider = dict(sorted(rider.items(), key = lambda d: d[1]['credit_score'], reverse = True))
        keys = list(rider.keys())
        if i >= rider[keys[j]]['finish_cur_delivery_time']:
            if len(rider[keys[j]]['cargo']) < 2:
                order_r_keys = list(order_r[i].keys())
                order_c_keys = list(order_c[i].keys())
                if len(order_r_keys) > 1:
                    for k in range(0, 2 - len(rider[keys[j]]['cargo'])):
                        rider[keys[j]]['cargo'].append({order_r_keys[k]: order_r[i].pop(order_r_keys[k]), 
                                                        order_c_keys[k]: order_c[i].pop(order_c_keys[k])})
                        rider[keys[j]]['start_cur_delivery_time'] = i
                elif len(order_r_keys) == 1:
                    rider[keys[j]]['cargo'].append({order_r_keys[0]: order_r[i].pop(order_r_keys[0]), 
                                                    order_c_keys[0]: order_c[i].pop(order_c_keys[0])})
                else:
                    continue
        else:
            continue
            
    # ------- put the rest of the order in that minute (if not taken by any rider) -------
    if len(order_r[i]) > 0:
        awaiting_order_r[i] = order_r[i]
        awaiting_order_c[i] = order_c[i]
        
    # ------- start delivering and updating rider status -------
    '''
    Here is where different delivery strategies will make a difference
    '''
    for j in range(0, 100):
        keys = list(rider.keys())
        if len(rider[keys[j]]['cargo']) == 2:
            des_1_keys = list(rider[keys[j]]['cargo'][0].keys())
            des_2_keys = list(rider[keys[j]]['cargo'][-1].keys())
            des_1_1 = rider[keys[j]]['cargo'][0][des_1_keys[0]]
            des_1_2 = rider[keys[j]]['cargo'][0][des_1_keys[1]]
            des_2_1 = rider[keys[j]]['cargo'][1][des_2_keys[0]]
            des_2_2 = rider[keys[j]]['cargo'][1][des_2_keys[1]]
            rider[keys[j]]['finish_cur_delivery_time'] = rider[keys[j]]['start_cur_delivery_time'] + \
                                                         (dis[rider[keys[j]]['cur_point'] - 1][des_1_1 - 1] + 
                                                          dis[des_1_1 - 1][des_1_2 - 1] + dis[des_1_2 - 1][des_2_1 - 1] +
                                                          dis[des_2_1 - 1][des_2_2 - 1]) / speed
            finished_orders[des_1_keys[0]] = [rider[keys[j]]['rider_number'], rider[keys[j]]['start_cur_delivery_time'], 
                                              rider[keys[j]]['start_cur_delivery_time'] + 
                                              (dis[rider[keys[j]]['cur_point'] - 1][des_1_1 - 1] + dis[des_1_1 - 1][des_1_2 - 1]) / speed]
            finished_orders[des_2_keys[0]] = [rider[keys[j]]['rider_number'], rider[keys[j]]['start_cur_delivery_time'], 
                                              rider[keys[j]]['finish_cur_delivery_time']] 
            if finished_orders[des_1_keys[0]][2] - int(des_1_keys[0].split('_')[1]) > 45:
                rider[keys[j]]['credit_score'] -= 1
            else:
                rider[keys[j]]['credit_score'] += 1
                
            if finished_orders[des_2_keys[0]][2] - int(des_2_keys[0].split('_')[1]) > 45:
                rider[keys[j]]['credit_score'] -= 1
            else:
                rider[keys[j]]['credit_score'] += 1
                
            rider[keys[j]]['cur_point'] = des_2_2
        else:
            continue
```
